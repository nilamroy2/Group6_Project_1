```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df_2016 = pd.read_csv("Medicare_Provider_Charge_Inpatient_FY2016.csv")

print(df_2016["Provider State"].nunique())
      
df_2016.head()
```

```python
df_2016['Average Covered Charges'] = df_2016['Average Covered Charges'].str.replace('$','').str.replace(',','')
df_2016['Average Covered Charges'] = df_2016['Average Covered Charges'].astype(np.float64)
```

```python
df_2016['Average Medicare Payments'] = df_2016['Average Medicare Payments'].str.replace('$','').str.replace(',','')
df_2016['Average Medicare Payments'] = df_2016['Average Medicare Payments'].astype(np.float64)
```

```python
df_2016["Average Total Payments"] = df_2016["Average Total Payments"].str.replace("," , "").str.replace("$" , "")
df_2016['Average Total Payments'] = df_2016['Average Total Payments'].astype(np.float64)
```

```python
new_df=df_2016[['DRG Definition','Provider Id','Provider State','Average Covered Charges','Average Medicare Payments','Average Total Payments']]
new_df.head()
new_df
```

```python
new_df['DRG Definition'].value_counts()
```

```python

new_df['Percentage_Medicare']=new_df['Average Total Payments'] - new_df['Average Medicare Payments']
new_df['Percentage_not_paid_by_Medicare']=(new_df['Percentage_Medicare']/new_df['Average Total Payments'])*100
#new_df['Percentage_not_paid_Medicare']=(new_df['Average Medicare Payments']/new_df['Average Covered Charges'])*100
new_df
```

```python
common_only = new_df.loc[new_df['DRG Definition'].isin(['293 - HEART FAILURE & SHOCK W/O CC/MCC',
                                                           '292 - HEART FAILURE & SHOCK W CC',
                                                           '194 - SIMPLE PNEUMONIA & PLEURISY W CC',
                                                           '684 - RENAL FAILURE W/O CC/MCC',
                                                           '699 - OTHER KIDNEY & URINARY TRACT DIAGNOSES W CC'])]
common_only
```

```python
new_df=common_only[["DRG Definition","Provider State","Percentage_not_paid_by_Medicare"]]
new_df
```

```python
new_df["DRG Definition"].value_counts()
```

```python
by_state_drg = common_only.groupby(['Provider State'],as_index=False,)['Percentage_not_paid_by_Medicare'].mean()
by_state_drg
```

```python
state_only = by_state_drg.loc[by_state_drg['Provider State'].isin(['CA','TX','FL','VT','ND'])]
state_only
```

```python
states = ["California", "Texas", "Florida", "Vermont", "North Dakota"]
bars_in_states = [17.465729, 15.183740, 37.345870, 18.074967, 46.665931]
x_axis = np.arange(len(bars_in_states))
```

```python
#Give the chart a title, x label, and y label
plt.title("Medicare coverage in different States")
#plt.xlabel("States")
plt.ylabel("Percentage NotPaid by Medicare(%)")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, states)
plt.bar(x_axis, bars_in_states, color="y", align="center")
plt.savefig("Different states coverage.png")
plt.show()
```
